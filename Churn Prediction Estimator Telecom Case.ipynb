**Training Preparation**

```python
x = dfset.drop(['international_plan','churn','voice_mail_plan','state','area_code'], axis=1)
y = dfset['churn']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report # performance measurement , untuk menghasilkan confusion matrix
from sklearn.svm import SVC
from sklearn import metrics
import matplotlib.pyplot as plt
import matplotlib
from sklearn import model_selection
import numpy as np
import math
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(x,y,test_size=0.25,shuffle=True,random_state=25)
```

Support Vector Machine

```python
model_svc = SVC(kernel='linear', random_state=0, probability = True)
model_svc.fit(x_train, y_train)
```

```python
pred_svc =model_svc.predict(x_test)
svc_acc = round(metrics.accuracy_score(y_test,pred_svc)*100,2)
print("Accuracy",svc_acc,"%")
```

```python
kfold = model_selection.KFold(n_splits=10,shuffle = True, random_state=30)
model = SVC()
results = model_selection.cross_val_score(model, x, y, cv=kfold,scoring='accuracy')
print("Accuracy: )", results*100)
print("Accuracy: %.3f%%" % (results.mean()*100.0))
```

```python
from sklearn.metrics import confusion_matrix

def print_confusion_matrix(confusion_matrix, class_names, figsize = (10,7), fontsize=14):
   
    df_cm = pd.DataFrame(
        confusion_matrix, index=class_names, columns=class_names, 
    )
    fig = plt.figure(figsize=figsize)
    try:
        heatmap = sns.heatmap(df_cm, annot=True, fmt="d")
    except ValueError:
        raise ValueError("Confusion matrix values must be integers.")
    heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right', fontsize=fontsize)
    heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=45, ha='right', fontsize=fontsize)
    plt.ylabel('Predict')
    plt.xlabel('Truth')
```

```python
cm = confusion_matrix(y_test, pred_svc)
print_confusion_matrix(cm,["no","yes"])
print(classification_report(y_test, pred_svc))
```

LOGISTIC REGRESSION

```python
num = {'yes': 1, 'no':0}
df = df.replace({'churn': num})
df = df.replace({'international_plan': num})
df = df.replace({'voice_mail_plan': num})
df.head( )
```

```python
feature_columns = ['international_plan', 'voice_mail_plan', 'Many_service_calls']
x_logreg = df[feature_columns]
y_logreg = df['churn']
```

```python
x_training, x_testing, y_training, y_testing = model_selection.train_test_split(x_logreg,y_logreg,test_size=0.25,shuffle=True,random_state=25)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
                   intercept_scaling=1, l1_ratio=None, max_iter=100,n_jobs=None, penalty='l2',
                   random_state=0, solver='liblinear', tol=0.0001, verbose=0,
                   warm_start=False)
logreg.fit(x_training,y_training)
```

```python
logreg.score(x_training,y_training)
```

```python
y_pred = logreg.predict(x_testing)
logreg_acc = round(metrics.accuracy_score(y_testing,y_pred)*100,2)
print("Accuracy",logreg_acc,"%")
```

```python
print("Classes : ", logreg.classes_)
print("Intercept : ",logreg.intercept_)
print("Coef : " ,logreg.coef_)
```

```python
logreg.predict_proba(x_training)
```
